# Exercício 1

Escreva uma função que receba uma lista com valores de receita obtidos em x dias e calcule a média móvel de três dias ao longo dos valores.

**Entrada**: Lista contendo valores de receita

**Saída**: Lista com média móvel

Exemplo:

Entrada: [1, 4, 6, 8, 10, 11, 15]

Operação:
* $\frac{1+4+6}{3}$
* $\frac{4+6+8}{3}$

Saída: [3.66, 6, 8, 9.66, 12]

```python
def gerar_dados_ma(size:int) -> list:
  """Gera dados do exercicio 1 e 2

  :param size: Número de observações
  :type size: int

  :return:
    lista: Lista com as observações
  """

  import random
  random.seed(15)

  if size <= 2:
    raise Exception('Tamanho deve ser no minimo 3')

  output_list = []
  output_list.append(random.gauss(0, 2))

  output_list.append(random.gauss(0, 2))
  for k in range(2, size):
    new_obs = 0.8*output_list[k-1] - 0.5*output_list[k-2] + random.gauss(0, 2)
    output_list.append(new_obs)

  return output_list
```

```python
#Para diferenciar as variáveis entre os exercicios usei e1, e2, e3, e4, e5 nos nomes
dados_gerados_e1=gerar_dados_ma(size=30) #10 dados por dia
dias_e1=3

def e1_funcao_media_movel(dados_e1):
  media_movel_e1=[]

  for a in range (len(dados_e1)):
    if a < 2:
      media_movel_e1.append(dados_e1[a])
    else:
      dados_trabalhados_e1=(dados_e1[a-2]+dados_e1[a-1]+dados_e1[a])/dias_e1
      media_movel_e1.append(round(dados_trabalhados_e1, 2))

  return media_movel_e1

media_movel_e1=e1_funcao_media_movel(dados_gerados_e1)

print('Entrada:\nDados da receita referente à 3 dias:')
print(dados_gerados_e1)
print('\nSaída:\nMédia móvel da receita (3 dias):')
print(media_movel_e1)
```

# Exercício 2
Escreva a função do exercício 1 porém com o parâmetro de dias da média móvel sendo um input.

**Entrada**: Dados, parâmetro da média móvel

**Saída**: Lista com as médias móveis

Observação: o código não deve funcionar caso o volume de dados seja menor que o dias necessários para a média móvel.

Exemplo:

Entrada: [1, 4, 6, 8, 10, 11, 15] e parâmetro = 4

Operação:
* $\frac{1+4+6+8}{4}$
* $\frac{4+6+8+10}{4}$

Saída: [4.75, 7, 8.75, 11]

```python
#Para diferenciar as variáveis entre os exercicios usei e1, e2, e3, e4, e5 nos nomes

dados_gerados_e2 = gerar_dados_ma(30)  
parametro = 4

def e2_funcao_media_movel_p(dados_e2, dias_parametro):
    if len(dados_e2) < dias_parametro:
        raise ValueError('O volume de dados da  deve ser maior do que os dias necessários para a média móvel.')
    
    media_movel_e2 = []
    
    for a in range(len(dados_e2)):
        if a < dias_parametro - 1:
            media_movel_e2.append(dados_e2[a])
        else:
            dados_trabalhados_e2 = (dados_e2[a - dias_parametro + 3]+dados_e2[a - dias_parametro + 2] + dados_e2[a - dias_parametro + 1] + dados_e2[a]) / dias_parametro
            media_movel_e2.append(round(dados_trabalhados_e2, 2))
    
    return media_movel_e2

media_movel_p_e2 = e2_funcao_media_movel_p(dados_gerados_e2, parametro)

print('Saída: Média móvel com input de dias')
print(media_movel_p_e2)
```

# Exercício 3
Desejamos realizar transformações em features a serem utilizadas em um modelo.

**Entrada**  receberemos uma lista com os valoers da feature

**Saída** uma lista com a feature; a feature utilizando a normalização z-score; feature utilizando a normalização maxmin


Z-score: $\frac{X - \bar{X}}{DP(X)}$.

$\bar{X}$ é média das features e $DP(X)$ é o desvio-padrão de x.

Normalização maxmin: $\frac{X-min(X)}{max(X)-min(X)}$

Exemplo:
Entrada: [1 , 2, 3]

Saída: [[1, 2, 3], [-1, 0, 1], [0, 0.5, 1]]

```python
def gerar_features(size: int) -> list:
  """Gera dados do exercicio 3

  :param size: Número de observações
  :type size: int
  :param features: Número de features
  :type features: int

  :return:
    lista: Dicionario com as observações
  """
  import random
  random.seed(15)

  if size <= 2:
    raise Exception('Tamanho deve ser no minimo 3')

  output_list = []
  for k in range(size):
    output_list.append(random.gauss(4, 3))

  return output_list
```

```python
#Para diferenciar as variáveis entre os exercicios usei e1, e2, e3, e4, e5 nos nomes
#Neste exercício usei a letra f para simbolizar a palavra "feature" nos nomes)

valores_f_e3 = gerar_features(size=5)

def e3_funcao_transformar_f(valores_f_e3):
    if len(valores_f_e3) <= 2:
        raise ValueError('Tamanho da lista de features deve ser no mínimo 3')

    media_e3 = sum(valores_f_e3) / len(valores_f_e3)
    dp_e3 = (sum((x - media_e3) ** 2 for x in valores_f_e3) / len(valores_f_e3)) ** 0.5

    f_original_e3 = []
    f_zscore_e3 = []
    f_maxmin_e3 = []

    for b in valores_f_e3:
        f_original_e3.append(b)

        z_score_e3 = (b - media_e3) / dp_e3
        f_zscore_e3.append(z_score_e3)

        valor_min_e3 = min(valores_f_e3)
        valor_max_e3 = max(valores_f_e3)
        maxmin_e3 = (b - valor_min_e3) / (valor_max_e3 - valor_min_e3)
        f_maxmin_e3.append(maxmin_e3)

    return [f_original_e3, f_zscore_e3, f_maxmin_e3]

features_e3 = e3_funcao_transformar_f(valores_f_e3)

print('Entrada:\nValores da feature')
print(valores_f_e3)
print('Saída:\nPrimeira lista: Features origonais; Segunda lista: Features com normalização z-score; Terceira lista: Features com normalização maxmin')
print(features_e3)
```

# Exercício 4
Desejamos realizar essa transformação em um dicionário de features, a transformação a ser utilizada sera um input na função. Se o input for z-score a normalização z-score deve ocorrer e caso seja max-min a normalização max-min deve ocorrer.

**Entrada** Dicionário com as features e tipo de padronização (string)

**Saída** Dicionário com a feature padronizada.

Exemplo:

Entrada:
{'Feature 1': [1, 2, 3],
'Feature 2': [1, 4, 10]}

Tipo de padronização: max-min

Saída:
{'Feature 1': [0, 0.5, 1],
'Feature 2': [0, 0.4, 1]}

```python
def gerar_features_dict(size: int, features: int) -> list:
  """Gera dados do exercicio 4

  :param size: Número de observações
  :type size: int
  :param features: Número de features
  :type features: int

  :return:
    lista: Dicionario com as observações
  """
  import random
  random.seed(15)

  if size <= 2:
    raise Exception('Tamanho deve ser no minimo 3')

  if features <= 1:
    raise Exception('Número de features deve ser no minimo 2')

  output_dict = {}
  mean = 3
  sd = 1

  for j in range(features):
    output_dict[f'feature_{j}'] = []
    for k in range(size):
      output_dict[f'feature_{j}'].append(random.gauss(mean, sd))
    mean += 0.5
    sd *= 1.1

  return output_dict
```

```python
#Para diferenciar as variáveis entre os exercicios usei e1, e2, e3, e4, e5 nos nomes
#Neste exercício usei a letra f para simbolizar a palavra "feature" nos nomes)

def e4_funcao_padronizar_f(dicionario_f_e4, padronizacao_e4):
    if padronizacao_e4 not in ['z-score', 'max-min']:
        raise ValueError('O tipo de padronização deve ser "z-score" ou "max-min".')

    dicionario_e4 = {}
    
    for c, valores_f_e4 in dicionario_f_e4.items():
        if padronizacao_e4 == 'z-score':
            
            media_e4 = sum(valores_f_e4) / len(valores_f_e4)
            dp_e4 = (sum((x - media_e4) ** 2 for x in valores_f_e4 / len(valores_f_e4)) ** 0.5)
            
            f_padrao_e4 = [(valor - media_e4) / dp_e4 for valor in valores_f_e4] #Normalização z-score

        elif padronizacao_e4 == 'max-min':

            valor_min_e4 = min(valores_f_e4)
            valor_max_e4 = max(valores_f_e4)

            f_padrao_e4 = [(b - valor_min_e4) / (valor_max_e4 - valor_min_e4) for b in valores_f_e4] #Normalização maxmin
        
        dicionario_e4[c] = f_padrao_e4
    
    return dicionario_e4

dicionario_f_e4 = gerar_features_dict(5,2)
padronizacao_e4 = 'max-min'
features_padronizadas_e4 = e4_funcao_padronizar_f(dicionario_f_e4, padronizacao_e4)

print('Entrada:\nFeatures e tipo de padronização')
print(dicionario_f_e4)
print('\nSaída:\nFeatures padronizadas')
print(features_padronizadas_e4)
```

# Exercício 5
Possuímos uma lista com o id do usuário e uma outra lista com o valor das compras realizadas por ele. Considerar que o usuário no indice 0 fez a compra do índice 0.

Supondo que o usuário aparece mais de uma vez criar um dicionário com chave id e valor máximo das compras

**Entrada**: Lista com id e lista com as compras

**Saída**: Dicionário com chave id do usuário e valor o máximo das compras

Exemplo:

Entrada:

Lista_id = [0, 1, 2, 1, 0]

Lista_compras = [100, 150, 200, 75, 300]

Saída:
{0: 300,
1: 150,
2: 200}


```python
from typing import Tuple
def gerar_listas(size: int, users: int) -> Tuple[list, list]:
  """Gera dados do exercicio 4

  :param size: Número de observações
  :type size: int
  :param users: Número de usuários
  :type features: int

  :return:
    Tuple[list, list]: Lista com o identificador dos usuários e lista com o valor da compra.
  """
  import random
  random.seed(15)

  if size <= users * 4:
    raise Exception('Amostra deve ser 4 vezes maior que o número de usuarios')



  id_user = []
  receita = []

  for k in range(size):
    id_user.append(random.randint(0, users))
    receita.append(random.gammavariate(10, 200))


  return id_user, receita
```

```python
#Para diferenciar as variáveis entre os exercicios usei e1, e2, e3, e4, e5 nos nomes

lista_gerada_e5 = gerar_listas(25,5)
chave_id_e5, val_max_compras_e5 = lista_gerada_e5 #seprar listas de id e compras

def e5_criar_dicionario(id_e5, compras_e5):
    dicionario_max_compras_e5 = {}
    
    for d in range(len(id_e5)):
        id_usuario_e5 = id_e5[d]
        valor_compra_e5 = compras_e5[d]
        
        if id_usuario_e5 in dicionario_max_compras_e5: #Checagem de id duplicado
            if valor_compra_e5 > dicionario_max_compras_e5[id_usuario_e5]: #Atualiza valor máximo
                dicionario_max_compras_e5[id_usuario_e5] = valor_compra_e5
        else:
            dicionario_max_compras_e5[id_usuario_e5] = valor_compra_e5 #Incluir id's novos
    
    return dicionario_max_compras_e5

dicionario_e5 = e5_criar_dicionario(chave_id_e5, val_max_compras_e5)

print('Entrada:\nLista com id e as compras')
print(chave_id_e5,val_max_compras_e5)
print('Saída:\nDicionário com chave id e valor máximo de compras')
print(dicionario_e5)
```
