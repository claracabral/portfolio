# Exercício 1


O dataset Titanic disponível no pacote seaborn contém dados dos passageiros do Titanic. Os campos presente no Dataset são:


1. survived: variável binária (1 se sobreviveu ou 0 contrario)
2. pclass: classe do ticket
3. sex: sexo do passageiro
4. age: idade em anos completos
5. sibsp: número de irmãos/cônjuges a bordo do navio
6. parch: número de pais/filhos a bordo do navio
7. fare: tarifa da passagem
8. embarked: porto de embarque (C=Cherbour, Q=Queenstown, S=Southampton)


Faça as análises a seguir:
1. Construa um boxplot com a distribuição da idade para os pacientes que sobreviveram e não sobreviveram. 
2. Construa um gráfico de barras que mostre o número de passageiros que sobreviveu e não sobreviveu por classe.
3. Construa um scatter plot mostrando a relação entre tarifa e idade; a cor de cada ponto deve ser referente a classe.

Dicas para as análises:

2. Para construir o gráfico construa uma tabela intermediária onde os índices são as classes e as colunas sobreviveu ou não.

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
#Análise 1
import matplotlib.pyplot as plt

sns.boxplot(x='survived', y='age', data=titanic, hue="survived")

plt.xlabel(
    "Não sobreviveram           Sobreviveram",
    size=16)
plt.ylabel("Idade", size=16)
plt.legend(labels=["Não sobreviveram (azul)", "Sobreviveram (laranja)"])
plt.title(
    "Gráfico da idade de passageiros do Titanic,\nconsiderando os que sobreviveram e que não sobreviveram",
    fontdict={"family": "serif", "color": "darkblue", "weight": "bold", "size": 18},
)
```

```python
#Criando tabela intermediária para Análise 2

sobrevivente_classe = pd.DataFrame(titanic.groupby('survived')['pclass'].value_counts())

sobrevivente_classe
```

```python
## Análise 2
import matplotlib.pyplot as plt

sns.barplot(sobrevivente_classe, x="pclass", y="count", hue="survived")

plt.xlabel(
    "Classe",
    size=16)
plt.ylabel("Número de passageiros", size=16)
plt.legend(labels=["Não sobreviveram (azul)", "Sobreviveram (laranja)"])
plt.title(
    "Gráfico de passageiros por classe,\nconsiderando os que sobreviveram e não sobreviveram",
    fontdict={"family": "serif", "color": "darkblue", "weight": "bold", "size": 18},
)

```

```python
## Análise 3
import matplotlib.pyplot as plt

sns.scatterplot(data=titanic, x="age", y="fare", hue="pclass")

plt.xlabel(
    "Idade",
    size=16)
plt.ylabel("Tarifa", size=16)
plt.legend(labels=["Classe 1", "Classe 2", "Classe 3"])
plt.title(
    "Gráfico da tarifa dos passageiros do Titanic por idade,\nconsiderando a classe de cada um",
    fontdict={"family": "serif", "color": "darkblue", "weight": "bold", "size": 18},
)
```
