# Exercício 1

O dataset Titanic disponível no pacote seaborn contém dados dos passageiros do Titanic. Os campos presente no Dataset são:

1. survived: variável binária (1 se sobreviveu ou 0 contrario)
2. pclass: classe do ticket
3. sex: sexo do passageiro
4. age: idade em anos completos
5. sibsp: número de irmãos/cônjuges a bordo do navio
6. parch: número de pais/filhos a bordo do navio
7. fare: tarifa da passagem
8. embarked: porto de embarque (C=Cherbour, Q=Queenstown, S=Southampton)


Faça as análises a seguir:
1. Calcule o percentual de sobreviventes para cada classe
2. Crie uma nova coluna com o seguinte intervalos de idade:
* Idade menores que 18 anos categorizados como 0 (idade < 18)
* Idade entre 18 e 30 anos categorizados como 1 (18 <= idade < 30)
* Idade entre 30 e 50 anos categorizados como 2 (30 <= idade < 50)
* Idade maior que 50 categorizados como 3 (50<= idade)
3. Calcule o percentual de sobrevivente de cada uma das categorias de idade

```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
## Analise 1


sobrevivente_classe = titanic.groupby('survived')['pclass'].value_counts(normalize=True).to_frame().sort_index()*100

print('Dos passageiros que sobreviveram, quantos % eram de cada classe?')
print('Não sobreviveram: 100% se unir todas as classes com survived = 0 \nSobreviveram: 100% se unir todas as classes com survived = 1')
print(sobrevivente_classe)
```

```python
## Analise 2

titanic['intervalo_idade'] = pd.cut(titanic['age'], bins = [0, 17, 29, 49, 100], labels = ['idade < 18','18 <= idade < 30','30 <= idade < 50','50 <= idade'])
interv_idade=titanic['intervalo_idade'].value_counts()

print(interv_idade)
```

```python
## Analise 3

sobrevivente_interv_idade=titanic.groupby('survived')['intervalo_idade'].value_counts(normalize=True).sort_index().to_frame()*100

print('Dos passageiros que sobreviveram, quantos % estavam nos intervalos de idade determinados?')
print('Não sobreviveram: 100% se unir todos os intervalos com survived = 0 \nSobreviveram: survived 1 = 100% se unir todos os intervalos com survived = 1')
print(sobrevivente_interv_idade)
```
