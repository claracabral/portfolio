Colunas:

```python
import pandas as pd
import numpy as np
import seaborn as sns # para o gráfico

from google.colab import drive
drive.mount('/content/drive')
```

## Fonte de dados: receita.csv

Este banco contém a receita de produtos comprados durante a visita de um usuário.

Colunas:

1. **id_visita**: Identificação da visita

2. **receita**: Receita gera pelos produtos comprados naquela visita.

```python
receita = pd.read_csv('/content/drive/MyDrive/dataset/receita.csv')
receita
```

# Tarefa 1

Unifique os dois dataframes em um utilizando o campo id_visita, preencha os dados
faltantes com valor 0.



```python
# novo é o DataFrame unificado sem eliminar as entradas NaN
novo = pd.merge(visitas, receita, on = 'id_visita', how = 'outer')
novo
```

```python
# bd é o dataframe unificado substituindo entradas NaN por zero
bd = novo.fillna(0)
bd
```

# Tarefa 2

Calcule as descritivas a seguir por id_cliente:

a. Média incluindo visitas sem receita

b. Média excluindo visitas sem receita

c. Percentual de visitas com receita


```python
# Resolvendo (a) usamos o DataFrame bd
visitasSemReceita = bd.groupby('id_cliente')['receita'].mean()
visitasSemReceita
```

```python
'''Resolvemos (b) usando o DataFrame antes de substituir as entradas NaN por zero
e removemos essas entradas para criar o DataFrame bd2'''
bd2 = novo.dropna()
visitasComReceita = bd2.groupby('id_cliente')['receita'].mean()
visitasComReceita
```

```python
# (c)
conta1 = bd2.groupby('id_cliente')['receita'].count() # quantidade de visitas com receita
conta2 = bd.groupby('id_cliente')['receita'].count() # quantidade total de visitas

conta1/conta2 * 100 # Percentual de visitas com receita em relação ao total de visitas
```

# Tarefa 3

Crie um algoritmo que faça os seguintes passos:
Para B indo de 1 até 1000:

1. Gere uma amostra (com reposição) do dataframe original. (df.sample(n=”tamanho do
banco”, replace=True))

2. Calcule a média de receita da amostra (considerando visitas sem receita)

3. Armazene esse valor em uma lista com o valor de B e uma lista com os valores de
média.
Criar um DataFrame utilizando essas listas.

4. Crie uma visualização para a distribuição da coluna media_amostrada no passo 3.


```python
num = []
medias = []
for i in range (1000):
   amostra = novo.sample(n=10, replace=True).fillna(0)
   num.append(i+1)
   medias.append(amostra.receita.mean())

df = pd.DataFrame ({
    "medias": pd.Series(medias),
    "num": pd.Series(num)
})

df.medias.plot.density()
```
